Посмотрим, насколько признаки скоррелированны друг с другом:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
corrmat = data.corr()
corrmat
```

Получили матрицу корреляций, посмотрим на нее в виде heatmap:

Посмотрим на еще одну визуализацию: seaborn.pairplot.

Эта функция выдает матрицу из графиков: по диагонали стоят гистограммы значений каждого признака, а в остальных ячейках — графики значений каждого признака против каждого. Кроме того, точки на графиках можно раскрасить в разные цвета по какому-либо категориальному признаку с помощью аргумента hue. Это позволяет сравнить, какие значения принимает признак в зависимости от категории. Давайте раскрасим наши графики по Voice mail plan, чтобы сравнить распределения признаков для не подключивших и подключивших пользователей.

```python
import warnings

warnings.filterwarnings('ignore')

sns.set()
cols = ['Account length', 'Area code', 'Number vmail messages', 'Voice mail plan', 'Total day minutes', 'Total day calls', 'Total intl minutes', 'Total intl calls', 'Churn']
sns.pairplot(data[cols], hue='Voice mail plan')
```

Построим отдельно самые интересные зависимости:

```python
sns.set()
cols = ['Area code', 'Number vmail messages', 'Voice mail plan', 'Total day minutes', 'Total day calls']
sns.pairplot(data[cols], hue='Voice mail plan')
```
